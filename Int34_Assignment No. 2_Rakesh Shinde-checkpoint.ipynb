# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
# To get the website
website="https://www.naukri.com/"
path="D:\chromedriver.exe"

#Open the Chrome window
driver=webdriver.Chrome(path)

#open the given website
driver.get(website)
driver.maximize_window()

# geting the input element
input_job=driver.find_element(By.XPATH,value="(//input[@type='text'])[1]")
input_location=driver.find_element(By.XPATH, value="(//input[@type='text'])[3]")
search_button=driver.find_element(By.XPATH, value="//div[@class='qsbSubmit']")

#sending keywords to scrap required data
input_job.send_keys("Data Analyst")
input_location.send_keys("Bangalore")
search_button.click()
sleep(3)

#Sraping first 10 jobs data

job_title=[]
job_location=[]
company_name=[]
experience_required=[]

# Scraping Job Titles
title=driver.find_elements(By.XPATH,value="//a[@class='title fw500 ellipsis']")
for i in title:
    job_title.append(i.text)
job_title=job_title[0:10]
job_title

#scrapping job locations
location=driver.find_elements(By.XPATH,value="//span[@class='ellipsis fleft fs12 lh16 locWdth']")
for i in location:
    job_location.append(i.text)
    
job_location=job_location[0:10]
job_location

#scraping company name
company=driver.find_elements(By.XPATH, value="//a[@class='subTitle ellipsis fleft']")
for i in company:
    company_name.append(i.text)
company_name=company_name[0:10]
company_name

#scraping experience required
experience=driver.find_elements(By.XPATH,value='//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
for i in experience:
    experience_required.append(i.text)
experience_required=experience_required[0:10]
experience_required

#creating dataframe for scrapped data
opportunities=pd.DataFrame({'Job Profile':job_title,'Job Location':job_location, 'Name of Company':company_name, 'Experience Required':experience_required})



driver.close()

opportunities
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# website tobe loaded
website="https://www.azquotes.com/"

#Webdriver Path
path='D:\chromedriver.exe'

# Creating Driver
driver=webdriver.Chrome(path)

#Loading Webpage
driver.get(website)

#maximize window
driver.maximize_window()
sleep(2)

#Click on the TopQuote button
button=driver.find_element(By.XPATH,value='/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
button.click()
sleep(2)

#scraping 1000 quotes with Quote, Author and type of Quote

#creating empty list
Quote=[]
Author=[]
Type=[]

#scraping data from 1st 9 pages

for k in range(9):

    for i in driver.find_elements(By.XPATH, value='//a[@class="title"]'):
        Quote.append(i.text)
    for i in driver.find_elements(By.XPATH, value='//div[@class="author"]'):
        Author.append(i.text)
    for i in driver.find_elements(By.XPATH, value='//div[@class="tags"]'):
        Type.append(i.text)
    next_button=driver.find_element(By.XPATH,value='(//li[@class="next"])[1]')
    next_button.click()
    sleep(4)
    
# To scrap last page

for i in driver.find_elements(By.XPATH, value='//a[@class="title"]'):
    Quote.append(i.text)
for i in driver.find_elements(By.XPATH, value='//div[@class="author"]'):
    Author.append(i.text)
for i in driver.find_elements(By.XPATH, value='//div[@class="tags"]'):
    Type.append(i.text)

#MAking Dataframe

Quotes_data=pd.DataFrame({'Quote':Quote,'Authors':Author, 'Type of Quote':Type})
 

driver.close()

Quotes_data
    
```
