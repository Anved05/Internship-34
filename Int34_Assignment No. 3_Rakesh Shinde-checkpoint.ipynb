# 1. Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Using selenium to scrap the items that are inputed by the user
# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.amazon.in/")

#Userinput

user_input=input('Please type the item you wants to search =')

# Entering the user input and click on search
driver.find_element(By.XPATH,"(//input[@class='nav-input nav-progressive-attribute'])[1]").send_keys(user_input)
driver.find_element(By.XPATH,"(//input[@class='nav-input nav-progressive-attribute'])[2]").click()


#creating empty lists for urls to scrape data individually visiting webpage
Product_urls=[]


# scrapping product urls from three pages
for i in range(0,3):
    links=driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-no-outline"]')  # scrapping Product urls for scrapping the data
    for i in links:
        Product_urls.append(i.get_attribute('href'))
    
    nxt_but=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_but.click()
    sleep(4)
        
Product_urls
```

```python
#creating empty lists for the data to be scraped respectively
Brand=[]
Product_Name=[]
Price=[]
Ret_Exch=[]
Exp_Delivery=[]
Availability=[]

for i in Product_urls:
    driver.get(i)
    try: 
        brand=driver.find_element(By.XPATH,'//table[@class="a-normal a-spacing-micro"]/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
    
    try:
        name=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product_Name.append(name.text)
    except NoSuchElementException:
        Product_Name.append('-')
    
    try:
        price=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[2]')
        Price.append(price.text)  
    except NoSuchElementException:
        Price.append('-')    
    
    try:
        ret=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        Ret_Exch.append(ret.text)
    except NoSuchElementException:
        Ret_Exch.append('-')
    
    try:
        delivery=driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')
        Exp_Delivery.append(delivery.text)
    except NoSuchElementException:
        Exp_Delivery.append('-')

    try:
        avail=driver.find_element(By.XPATH,'//*[@id="availability"]/span')
        Availability.append(avail.text)
    except NoSuchElementException:
        Availability.append('-')
    sleep(1)
    
# Making DataFrame for the scrapped data  

data=pd.DataFrame({'Brand':Brand,'Name of Product':Product_Name,'Price':Price,'Return/Exchange policy':Ret_Exch,'Expected Delivery':Exp_Delivery,'Availability':Availability,'Product URL':Product_urls})

driver.close()

data
```

# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
# Using selenium to scrap the items that are inputed by the user
# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.flipkart.com/")

# Close popup to sign in
driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()

#Userinput
user_input=input('Please type the item you wants to search =')

# Entering the user input and click on search
driver.find_element(By.XPATH,'//input[@class="_3704LK"]').send_keys(user_input)
driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]').click()
sleep(5)


#scraping urls of the given search item (One Plus Nord)
URL=[]

url=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')

for i in url:
    URL.append(i.get_attribute('href'))
    
```

```python
# Scrapping the data from 1st page respectively
Brand_Name=[] 
Name=[]
Colour=[]
RAM=[]
ROM=[]
Pri_Cam=[]
Sec_Cam=[]
Display=[]
Battery=[]
Price=[]

for i in URL:
    driver.get(i)
    driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]').click()
    sleep(1) #clicking on specification
    
    # scrapping Brand name
    brand=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
    for i in brand:  
        brd=i.text
    x=brd.split()
    Brand_Name.append(x[0])
    
    #scrapping model name
    try:
        name=driver.find_elements(By.XPATH,'(//li[@class="_21lJbe"])[3]')
        for i in name:
            Name.append(i.text)  
    except NoSuchElementException:
        Name.append('-')
   
    # scrapping RAM 
    try:
        ram=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[1]')
        for i in ram:
            y= i.text.split('|')[0]
        RAM.append(y)
    except NoSuchElementException:
        RAM.append('-') 
    
    # scrapping Colour
    try:
        colour=driver.find_element(By.XPATH,'(//li[@class="_21lJbe"])[4]')
        Colour.append(colour.text)  
    except NoSuchElementException:
        Colour.append('-')
    
    # scrapping ROM
    try:
        rom=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[1]')
        for i in rom:    
            x=i.text.split('|')[1]
        ROM.append(x)
    except NoSuchElementException:
        ROM.append('-')

    #scrapping Primary camera
    try:
        pc=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[3]')
        for i in pc:
            t= i.text.split('|')[0]
        Pri_Cam.append(t)
    except NoSuchElementException:
        Pri_Cam.append('-')
    
    #scrapping Secondary cam
    try:
        sc=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div/div/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[6]/td[2]/ul/li')
        Sec_Cam.append(sc.text)
    except NoSuchElementException:
        Sec_Cam.append('-')

    # scrapping display details
    try:
        display=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[2]')
        Display.append(display.text)
    except NoSuchElementException:
        Display.append('-')

    # scrapping battery details
    try:
        bat=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[6]/div[1]/div/div[2]/ul/li[4]')
        Battery.append(bat.text)
    except NoSuchElementException:
        Battery.append('-')
        
    #scrapping price
    try:
        price=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div[3]/div[1]/div/div')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    
    sleep(10)
        
# Making Data frame for thr data scrapped

df=pd.DataFrame({'Brand name':Brand_Name,'Model name':Name,'Colour':Colour,'RAM':RAM,'External Memory - ROM':ROM,'Primary Camera':Pri_Cam,'Secondary Camera':Sec_Cam,'Display':Display,'Battery':Battery,'Price':Price,'Product URL':URL})

driver.close()
df
```

# 8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#importing ActionChains Class
from selenium.webdriver import ActionChains

# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.forbes.com/?sh=5c9312f2254c")
sleep(2)

# clicking on list button
button1=driver.find_element(By.XPATH,'//div[@class="_69hVhdY4"]')
button1.click()

# calling actionchain method
actionchain=ActionChains(driver)

# hovering over the button on billionairs
button2=driver.find_element(By.XPATH,'(//li[@class="TjJgrPSg cD45ib6e primary"])[1]')
sleep(4)
# moving mouse on that web element
actionchain.move_to_element(button2).perform()
sleep(2)

# clicking on billionaires
button3=driver.find_element(By.XPATH,'(//li[@class="TjJgrPSg _2bNo56RE secondary"])[1]')
button3.click()
sleep(10)

# scraping the details from each page
# creating empty lists 
Rank=[]
Name=[]
NetWorth=[]
Age=[]
Country=[]
Source=[]
Industry=[]

# creating range for the data to be scrapped
start=1
end=int(input('data upto page : ')) # one can enter for the number of pages tobe scrapped and rhe data is available 200 per page

for k in range(start,end):
    
    for i in driver.find_elements(By.XPATH,'//div[@class="rank"]'):
        Rank.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="personName"]'):
        Name.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="netWorth"]'):
        NetWorth.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="age"]'):
        Age.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]'):
        Country.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="source"]'):
        Source.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="category"]'):
        Industry.append(i.text)
        
    # Going to next page
    nextbutton=driver.find_element(By.XPATH,'(//span[@class="bubble-arrow__label"])[2]')
    nextbutton.click()
    sleep(5)

Billionaires=pd.DataFrame({'Rank':Rank,'Name':Name,'Net worth':NetWorth,'Age':Age,'Citizenship':Country,'Source':Source,'Industry':Industry})

Billionaires

driver.close()

Billionaires
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.youtube.com/watch?v=kJQP7kiw5Fk")
sleep(20)

# Scraping the dtails

# Scrolling page

for _ in range(100):

    driver.execute_script("window.scrollBy(0,500)")
    
    sleep(5)
    
    Commentor=[]
    commentor=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]/h3/a')
    for i in commentor:
        Commentor.append(i.text)
    
    Comment=[]
    comment=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
    for i in comment:
        Comment.append(i.text)

    Time=[]
    time=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in time:
        Time.append(i.text)

    Upvote=[]
    upvote=driver.find_elements(By.ID,"vote-count-middle")
    for i in upvote:
        Upvote.append(i.text)

    
# Making a DataFrame for the scrapped information

data=pd.DataFrame({'Author':Commentor,'Comment':Comment,'Posting Time':Time,'Upvotes':Upvote})


driver.close()

data
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.hostelworld.com/")
sleep(20)

# to scroll down in the given page
driver.execute_script("window.scrollBy(0,8000)")
sleep(5)

# to click on London so as to scrape the details of hostel

element=driver.find_element(By.XPATH,'(//div[@class="mosaic-inner"])[1]/a[1]')
driver.get(element.get_attribute('href'))
sleep(10)

# Scrapping required information
# Creating empty list
Name=[]
Distance=[]
Rating=[]
Total_Reviews=[]
Overall_Review=[]
Price=[]
Facility=[]
Description=[]

# Scrapping information from all 4 pages

for page in range(0,4):
    
    for i in driver.find_elements(By.XPATH,'//h2[@class="title"]/a'): #scrapping name of hostel
        Name.append(i.text)

    for i in driver.find_elements(By.XPATH,'//span[@class="description"]'): # scrapping distance from city centre
        Distance.append(i.text)

    rat=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')    
    if driver.find_elements(By.XPATH,'//div[@class="score orange big"]'): # to scrap the rating
        for i in rat: 
            Rating.append(i.text)
    else:
        Rating.append('-')

    tr=driver.find_elements(By.XPATH,'//div[@class="rating-summary-container big"]/div[2]/div[2]')
    if driver.find_elements(By.XPATH,'//div[@class="rating-summary-container big"]/div[2]/div[2]'):  # to scrap the total review
        for i in tr:         # scrapping Total Reviews
            Total_Reviews.append(i.text)
    else:
        Total_Reviews.append('-')

    ovr=driver.find_elements(By.XPATH,'//div[@class="rating-summary-container big"]/div[2]/div[1]')    
    if driver.find_elements(By.XPATH,'//div[@class="rating-summary-container big"]/div[2]/div[1]'):  # to scrap the overall review
        for i in ovr: # scrapping Overall review
            Overall_Review.append(i.text)
    else:
        Overall_Review.append('-')

    for i in driver.find_elements(By.XPATH,'//a[@class="prices-row"]/span[2]'): # Scraping price from
        Price.append(i.text)

    for i in driver.find_elements(By.XPATH,'//div[@class="facilities"]'): # scrapping Facilities
        Facility.append(i.text.replace('\n', ' , '))
    
    for i in driver.find_elements(By.XPATH,'//p[@class="intro-text"]'): # scrapping hostel description
        Description.append(i.text)
    sleep(4)
    
    # Clicking on next page
    try:
        next_button=driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()  
        sleep(5)
    except NoSuchElementException:
        break

# to put '-' in place where the Rating and Review is not present
k=76
for k in range(76,93,1):
    Rating.insert(k,'-')
    Total_Reviews.insert(k,'-')
    Overall_Review.insert(k,'-')
    
Hostel_Data=pd.DataFrame({'Hostel Name': Name,'Distance from City centre': Distance,'Rating':Rating,'Total Reviews':Total_Reviews,'Overall Review':Overall_Review,'Price From':Price,'Facility': Facility,'Description':Description})

driver.close()

Hostel_Data
```
