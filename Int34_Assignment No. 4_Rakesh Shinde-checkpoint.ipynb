# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#loading website
driver=webdriver.Chrome('D:\chromedriver.exe') # creating automated chrome window
driver.maximize_window() # maximizing
driver.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos") # going to website
sleep(2)

# Preparing empty lits to scrape data
Rank=[]
Video=[]
Artist=[]
Date=[]
Views=[]

# Scrapping the required data
#scrapping rank
rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank[0:30]:
    Rank.append(i.text)

#scrapping name of video
name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in name[0:30]:
    video_name=i.text
    Video.append(re.findall(r'"(.*?)"', video_name))

#scrapping Name of uploader
artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in artist[0:30]:
    Artist.append(i.text)

#scrapping  date of upload
date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in date[0:30]:
    Date.append(i.text)

#scrapping no. of views
views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in views[0:30]:
    Views.append(i.text)

# making a dataframe
Video_list=pd.DataFrame({'Rank':Rank,'Video name': Video,'Name of Uploader':Artist,'date of Upload':Date,'Views(in Billions)': Views})

driver.close()

Video_list
```

# 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/

```python
#loading website
driver=webdriver.Chrome('D:\chromedriver.exe') # creating automated chrome window
driver.maximize_window() # maximizing
driver.get("https://github.com/") # going to website
sleep(2)

# clicking on open source button
button=driver.find_element(By.XPATH,'(//button[@class="HeaderMenu-link border-0 width-full width-lg-auto px-0 px-lg-2 py-3 py-lg-2 no-wrap d-flex flex-items-center flex-justify-between js-details-target"])[3]')
button.click()

# Clicking on trending repository
driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a').click()
```

```python
#Creating the lists of repository urls
URL=[]
url=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]/a')
for i in url:
    URL.append(i.get_attribute('href'))
```

```python
#scrapping the details by going on eachpage
#creating empty lists
Title=[]
Description=[]
Count=[]
Lang=[]

#Scrapping information from each repository
for i in URL:
    driver.get(i)
    sleep(1)

    #Scrapping the title
    title=driver.find_element(By.XPATH,'//div[@class="flex-auto min-width-0 width-fit mr-3"]')
    Title.append(title.text.replace('\n', ''))

    #Scrapping the description
    description=driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
    Description.append(description.text)

    #Scrapping the Contributer's count
    try:
        count=driver.find_element(By.XPATH,'(//span[@class="Counter"])[10]')
        Count.append(count.text)
    except NoSuchElementException:
        Count.append('-')
    
    #Scrapping the language used
    try:
        lang=driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        Lang.append(lang.text)
    except NoSuchElementException:
        Lang.append('-')


#making the data frame
Repository_Data=pd.DataFrame({'Repository Title':Title,'Repository Description':Description,'contributers Count':Count,'Language used':Lang})

Repository_Data
```

```python
#scrapping the required information
# creating empty lists
Name=[]
Designation=[]
Company_Name=[]
Skills=[]
Location=[]

#scrapping Recruiter's name
recruit=driver.find_elements(By.XPATH,'//span[@class="fl ellipsis"]')
for i in recruit:
    Name.append(i.text)

#scrapping Recruiter's designation
desg=driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
for i in desg:
    Designation.append(i.text)

#scrapping Recruiter's Company
Company=[]
company=driver.find_elements(By.XPATH,'//a[@class="ellipsis"]')
for i in company:
    Company.append(i.text)
for i in range(1,100,2):
    Company_Name.append(Company[i])

#scrapping skills they hire for
skills=driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
for i in skills:
    Skills.append(i.text)

#scrapping Company Location
location=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/span[2]/small')
try:
    for i in location:
        Location.append(i.text)
except NoSuchElementException:
    Location.append('Not Mentioned')

Location.insert(30,'Not Found') # inserting not found at the required places beacause try and except is not givig exception though exception handled
Location.insert(49,'Not Found')

# making a dataframe
Recruiter_data=pd.DataFrame({'Name of Recruiter':Name,'Designation of Recruiter':Designation,'Name of Company':Company_Name,'Skills They Hire for':Skills,'Location':Location})

driver.close()

Recruiter_data
```

B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the ShowAllDataset page through code.

```python
#loading website
driver=webdriver.Chrome('D:\chromedriver.exe') # creating automated chrome window
driver.maximize_window() # maximizing
driver.get("https://archive.ics.uci.edu/") # going to website
sleep(10)

# click on to all view all dataset
driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td/span/b/a').click()
sleep(10)

#scrapping weblink
weblinks=[]

#scraping weblinks
for i in driver.find_elements(By.XPATH,'//p[@class="normal"]/b/a'):
    weblinks.append(i.get_attribute('href'))
    
```

```python
#scrapping information 
# creating empty lists
Name=[]
Type=[]
Task=[]
Att_Type=[]
Instance=[]
Attributes=[]
Year=[]
for i in weblinks:
    try:
        driver.get(i)
    # Scrapping name of dataset
        name=driver.find_element(By.XPATH,'//span[@class="heading"]')
        Name.append(name.text)

    # Scrapping Type of dataset
        tipe=driver.find_element(By.XPATH,'(//p[@class="normal"])[3]')
        Type.append(tipe.text)

    # Scrapping Task
        task=driver.find_element(By.XPATH,'(//p[@class="normal"])[15]')
        Task.append(task.text)
    
    # Scrapping Attribute type
        att=driver.find_element(By.XPATH,'(//p[@class="normal"])[9]')
        Att_Type.append(att.text)

    # Scrapping Instance
        instance=driver.find_element(By.XPATH,'(//p[@class="normal"])[5]')
        Instance.append(instance.text)

    # Scrapping Attributes
        attributes=driver.find_element(By.XPATH,'(//p[@class="normal"])[11]')
        Attributes.append(attributes.text)

    # Scrapping Year
        year=driver.find_element(By.XPATH,'(//p[@class="normal"])[13]')
        Year.append(year.text)
    
    except:
        pass
# making a dataframe
dataset=pd.DataFrame({'Name of Dataset':Name,'Type of Dataset':Type,'Task':Task,'Attribute type':Att_Type,'No. of Instances':Instance,'No.of Attributes':Attributes,'Year':Year})

dataset
```
